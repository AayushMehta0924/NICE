```python
from tensorflow.keras.preprocessing.image import img_to_array, load_img
import matplotlib.pyplot as plt
from keras.models import load_model
import numpy as np
import tensorflow as tf

# Get the prediction
img = "/home/sky/Desktop/god-classification/k.jpg"
img = load_img(img, target_size=(412, 412))
plt.imshow(img)
img = img_to_array(img)
img = np.expand_dims(img, axis=0)

model = load_model("/home/sky/Desktop/god-classification/res_model10.h5")

class_ = np.argmax(model.predict(img,verbose=0))

predictions = model.predict(img,verbose=0)
score = tf.nn.sigmoid(predictions[0])

if class_ == 0:
    print("Brihaspati God")
elif class_ == 1:
    print("Ketu God")
else:
    print("Surya God")    
    
print("Accuracy : {:.2f} percent confidence.".format(100 * np.max(score)))
```
